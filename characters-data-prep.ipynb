```python
df['mass'] = df['mass'].str.replace(',', '').astype(float)
```

```python
df.dtypes
```

```python
avg_height = df["height"].astype("float").mean(axis=0)
df["height"].replace(np.nan, avg_height, inplace=True)
```

```python
avg_mass = df["mass"].astype("float").mean(axis=0)
df["mass"].replace(np.nan, avg_mass, inplace=True)
```
